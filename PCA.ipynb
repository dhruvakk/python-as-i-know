```
data=pd.read_csv("mtcars.csv")
data.head()
```

```
data= data.iloc[:,1:]
```

```
from sklearn.preprocessing import scale, normalize
```

```
data1 = scale(data)
```

```
data2= normalize(data1)
```

```
from sklearn.decomposition import PCA
```

```
model= PCA()
```

```
model.components_
```

```
component= pd.DataFrame(model.components_)
```
